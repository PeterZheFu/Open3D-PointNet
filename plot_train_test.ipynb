```python
%matplotlib inline
import pandas as pd
```

```python
N = 400
```

# Train

```python
log_xyz_train = pd.read_csv('log_xyz_train.csv')
log_xyz_train
```

```python
log_xyz_train = pd.read_csv('log_xyz_train.csv')
a = log_xyz_train.iloc[0:N]
a
```

```python
log_xyzs_train = pd.read_csv('log_xyzs_train.csv')
b = log_xyzs_train.iloc[0:N]['4']
```

```python
a['5'] = b
```

```python
a.rename(columns={ a.columns[0]: "number of epochs" }, inplace=True)
a.rename(columns={ a.columns[5]: "xyz" }, inplace=True)
a.rename(columns={ a.columns[6]: "xyz-semantic" }, inplace=True)
```

```python
a.plot(x='number of epochs', y=['xyz', 'xyz-semantic'], figsize=(15,10), grid=True, title='training for 400 epoch')
```
